```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt 

import warnings
warnings.filterwarnings('ignore')
```

```python
# us_locations = country_names + state_names + city_names
apple_mobility = pd.read_csv('data/applemobilitytrends-2020-04-27.csv') # will need to change this
apple_countries = apple_mobility[apple_mobility['geo_type'] == 'country/region']
# apple_us = apple_mobility[apple_mobility['region'].isin(us_locations)]
```

```python
greece = apple_countries[apple_countries.region == 'Greece']
greece
```

```python
# using nicks code to plot mobility in Greece
trans_type = ['driving','walking']

num_days = len(greece.columns[4:])
base = datetime.datetime(2020, 1, 13)
date_list = [base + datetime.timedelta(days=x) for x in range(num_days)]

plt.figure(0, figsize=(10,5))
for i in range(2):
    y = greece.values[i][4:]
    plt.plot(date_list, y, label=trans_type[i])
plt.legend()
plt.title('Mobility in Greece')
plt.xlabel('Date')
plt.ylabel('Percentage Mobility')
plt.show()
```

```python
usa = apple_countries[apple_countries.region == 'United States']

# using nicks code to plot mobility 
trans_type = ['driving','transit','walking']

num_days = len(usa.columns[4:])
base = datetime.datetime(2020, 1, 13)
date_list = [base + datetime.timedelta(days=x) for x in range(num_days)]

plt.figure(0, figsize=(10,5))
for i in range(3):
    y = usa.values[i][4:]
    plt.plot(date_list, y, label=trans_type[i])
plt.legend()
plt.title('Mobility in Greece')
plt.xlabel('Date')
plt.ylabel('Percentage Mobility')
plt.show()
```
